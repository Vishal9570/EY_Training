```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("sakshi_ppg_20260611T074737_len148s.csv")

# Mean values of signals
mean_values = {
    "Red": df["red"].mean(),
    "IR": df["ir"].mean(),
    "Red Corrected": df["red_corrected"].mean(),
    "IR Corrected": df["ir_corrected"].mean()
}

# Bar Plot
plt.figure(figsize=(8,5))
plt.bar(mean_values.keys(), mean_values.values())
plt.title("Average Signal Values")
plt.ylabel("Mean Value")
plt.xlabel("Signal Type")
plt.xticks(rotation=15)
plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv(
    r"C:\Users\Administrator\Learnings\Day11\sakshi_ppg_20260611T074737_len148s.csv"
)

# Use timestamp as index
df['timestamp_ms'] = pd.to_datetime(df['timestamp_ms'], unit='ms')
df.set_index('timestamp_ms', inplace=True)

# Decompose red signal
result = seasonal_decompose(
    df['red'],
    model='additive',
    period=100   # adjust according to your signal cycle
)

result.plot()
plt.show()
```
